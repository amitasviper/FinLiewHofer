```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (16, 9)
plt.style.use('ggplot')
```

```python
X = np.linspace(-2*np.pi, 2*np.pi, 100, endpoint=True)
y_sin = np.sin(X)
y_cos = np.cos(X)
```

```python
plt.plot(X, y_sin, 'b--', label="Sin")
plt.plot(X, y_cos, label="Cosine")
plt.xlabel('Angle in radians (x)')
plt.ylabel('Sin(x)')
plt.legend(loc='upper right')
plt.ylim(-1.5, 2.0)
plt.show()
```
